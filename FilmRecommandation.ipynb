```python
#!pip install sparqlwrapper
# https://rdflib.github.io/sparqlwrapper/

import pandas as pd
import sys
from SPARQLWrapper import SPARQLWrapper, JSON

endpoint_url = "https://query.wikidata.org/sparql"

query = """SELECT DISTINCT ?item ?itemLabel WHERE {
  ?item wdt:P31 wd:Q11424.
  ?item wdt:P577 ?pubdate.
  FILTER((?pubdate >= "2010-01-01T00:00:00Z"^^xsd:dateTime) && (?pubdate <= "2022-12-31T00:00:00Z"^^xsd:dateTime))
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}"""


def get_results(endpoint_url, query):
    user_agent = "WDQS-example Python/%s.%s" % (sys.version_info[0], sys.version_info[1])
    # TODO adjust user agent; see https://w.wiki/CX6
    sparql = SPARQLWrapper(endpoint_url, agent=user_agent)
    sparql.setQuery(query)
    sparql.setReturnFormat(JSON)
    return sparql.query().convert()


results = get_results(endpoint_url, query)

# Initialiser des listes pour stocker les données
items = []
item_labels = []

# Parcourir les résultats et extraire les données
for result in results["results"]["bindings"]:
    items.append(result["item"]["value"])
    item_labels.append(result["itemLabel"]["value"])

# Créer une DataFrame à partir des listes de données
df = pd.DataFrame({"Item": items, "ItemLabel": item_labels})

# Afficher la DataFrame
df.head(10)
```
